```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
from datetime import datetime
```

```python
base_url = "https://premierleague.com/match/"
```

```python
first_match_id = "5567" # should be 5567 for start of 06/07 season
example_url = base_url + first_match_id
response = requests.get(example_url,headers={'User-Agent': 'Custom'})
page = soup(response.content, "html.parser")
```

```python
json_data = page.find("div",{"class":"mcTabsContainer"})["data-fixture"]
json_data
```

```python
import json
with open('mydata.json', 'w') as f:
    json.dump(json_data, f)
```

```python
with open('mydata_indented.json', 'w') as f:
    json.dump(json_data, f, indent=4, separators=(",", ":"))
```

```python
with open("mydata_indented.json", "r") as read_file:
    data = json.load(read_file)
```

```python
print(json.dumps(data,indent=4))
```

```python
with open('mydata_indented.json') as f:
    decoded_once = json.load(f)
    new_data = json.loads(decoded_once)
```
