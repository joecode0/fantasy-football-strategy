```python
import pandas as pd
import ast
import time
```

```python
df = pd.read_csv("all_fpl_points.csv")
df_names = pd.read_csv("all_player_ids_and_names.csv")
```

```python
df.head()
```

```python
s = "19/20"
df2 = df.query("season == @s")
df3 = df2[["player_id","pos","points"]]
df3.head()
```

```python
df_names.head()
```

```python
dfn = df_names
d = dict([(i,a) for i, a in zip(dfn.id, dfn.name)])
print(d)
```

```python
def get_player_name(row):
    player_id = row["player_id"]
    name = d.get(int(player_id))
    return name
```

```python
df3["name"] = df3.apply(get_player_name,axis=1)
df3.head()
```

```python
df4 = df3[["player_id","pos","name","points"]]
df4.head()
```

```python
def total_season_points(df):
    output = []
    unique_player_ids = set(df["player_id"].tolist())
    for player in unique_player_ids:
        df_player = df.query("player_id == @player")
        top_row = df_player.iloc[0]
        pos = str(top_row["pos"])
        name = str(top_row["name"])
        total_points = sum(df_player["points"].tolist())
        output_row = [player,pos,name,total_points]
        output.append(output_row)
    output_df = pd.DataFrame(data=output,columns=["player_id","pos","name","total_points"])
    return output_df
```

```python
total_points_df = total_season_points(df4)
total_points_df.head()
```

```python
total_points_df_sorted = total_points_df.sort_values(["name"],axis=0)
```
