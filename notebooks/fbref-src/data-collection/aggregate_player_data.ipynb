```python
def aggregate_player_data(df):
    team_name_tuples = list(set(df.copy().set_index(['team','name']).index.tolist()))

    list_of_series = []
    
    for i in range(len(team_name_tuples)):
        t = team_name_tuples[i]
        team = t[0]
        name = t[1]
        df_tuple = df.query("team == @team & name == @name")
        data_dict = combine_player_data(df_tuple,team,name)
        data_series = pd.Series(data_dict)
        list_of_series.append(data_series)
        
    output_df = pd.DataFrame(list_of_series)
    return output_df
```

```python
def combine_player_data(df,team,name):
    df.fillna(0,inplace=True)
    first_row = df.iloc[0]
    d = {}
    d["team"] = str(team)
    d["name"] = str(name)
    d["matches"] = int(len(df))
    d["pos"] = str(first_row["pos"])
    d["age"] = int(first_row["age"])
    sum_list = ["minutes","goals","assists","pens_made","pens_att","shots_total","shots_on_target","cards_yellow","cards_red",
               "touches","pressures","tackles","interceptions","blocks","xg","npxg","xa","sca","gca","passes_completed",
               "passes","passes_progressive_distance","carries","carry_progressive_distance","dribbles_completed","dribbles",
               "passes_total_distance","passes_completed_short","passes_short","passes_completed_medium","passes_medium",
               "passes_completed_long","passes_long","assisted_shots","passes_into_final_third","passes_into_penalty_area",
               "crosses_into_penalty_area","progressive_passes","passes_live","passes_dead","passes_free_kicks","through_balls",
               "passes_pressure","passes_switches","crosses","corner_kicks","corner_kicks_in","corner_kicks_out",
               "corner_kicks_straight","passes_ground","passes_low","passes_high","passes_left_foot","passes_right_foot",
               "passes_head","throw_ins","passes_other_body","passes_offsides","passes_oob","passes_intercepted",
               "passes_blocked","tackles_won","tackles_def_3rd","tackles_att_3rd","dribble_tackles","dribbles_vs",
               "dribbled_past","pressure_regains","pressures_def_3rd","pressures_mid_3rd","pressures_att_3rd","blocked_shots",
                "blocked_shots_saves","blocked_passes","tackles_interceptions","clearances","errors","touches_def_pen_area",
               "touches_def_3rd","touches_mid_3rd","touches_att_3rd","touches_att_pen_area","touches_live_ball",
               "players_dribbled_past","nutmegs","carry_distance","pass_targets","passes_received","miscontrols","dispossessed",
               "fouls","fouled","pens_won","pens_conceded","own_goals","ball_recoveries","aerials_won","aerials_lost"]
    
    
    rename_list = ["total_mins","goals","assists","pens_made","pens_attempted","shots","shots_on_target","yellow_cards",
                  "red_cards","touches","pressures","tackles","interceptions","blocks","xg","npxg","xa","sca","gca",
                  "passes_completed","passes_attempted","passes_prog_dist","carries","carries_prog_dist","dribbles_completed",
                  "dribbles_attempted","passes_total_dist","passes_completed_short","passes_attempted_short",
                  "passes_completed_medium","passes_attempted_medium","passes_completed_long","passes_attempted_long",
                  "assisted_shots","passes_into_final_third","passes_into_penalty_area","crosses_into_penalty_area",
                  "prog_passes","passes_live","passes_dead","passes_free_kicks","through_balls","passes_under_pressure",
                  "switches","crosses","corner_kicks","corner_kicks_in","corner_kicks_out","corner_kicks_straight",
                  "passes_ground","passes_low","passes_high","passes_left_foot","passes_right_foot","passes_head","throw_ins",
                  "passes_other_body","passes_offside","passes_out_of_bounds","passes_intercepted","passes_blocked",
                  "tackles_won","tackles_def_3rd","tackles_att_3rd","dribblers_tackles","dribblers_attempted_vs",
                  "dribbled_past","pressure_regains","pressures_def_3rd","pressures_mid_3rd","pressures_att_3rd",
                  "blocked_shots","blocked_shots_on_target","blocked_passes","tackles_interceptions","clearances",
                  "errors","touches_def_pen_area","touches_def_3rd","touches_mid_3rd","touches_att_3rd","touches_att_pen_area",
                  "touches_live_ball","players_dribbled_past","nutmegs","carry_distance","pass_target_count","passes_received",
                  "miscontrols","dispossessed","fouls","times_fouled","pens_won","pens_conceded","own_goals","ball_recoveries",
                  "aerials_won","aerials_lost"]
    
    count = len(sum_list)
    if count == len(rename_list):
        for i in range(count):
            key = rename_list[i]
            d[key] = int(sum(df[sum_list[i]].tolist()))
    else:
        print("Different lengths, fix before continuing!")
        return {}
        
    return d  
```
