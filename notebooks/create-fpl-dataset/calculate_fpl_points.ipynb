```python
import pandas as pd
import time
import ast
df = pd.read_csv("data/all_fpl_data.csv")
df.head()
```

```python
df.drop("del",inplace=True,axis=1)
df.head()
```

```python
def calculate_fpl_points(row):
    total_points = 0
    played_points = calculate_played_points(int(row["mins"]))
    total_points += played_points
    
    goals_points = calculate_goals_points(int(row["goals"]),str(row["pos"]))
    total_points += goals_points
    
    assist_points = calculate_assist_points(int(row["assists"]))
    total_points += assist_points
    
    clean_sheet_points = calculate_clean_sheet_points(str(row["clean_sheet"]),str(row["pos"]))
    total_points += clean_sheet_points
    
    penalty_save_points = calculate_pen_save_points(int(row["pen_saves"]),str(row["pos"]))
    total_points += penalty_save_points
    
    penalty_miss_points = calculate_pen_miss_points(int(row["pen_misses"]))
    total_points -= penalty_miss_points
    
    goals_conceded_points = calculate_goals_conceded_points(int(row["goals_conceded"]),str(row["pos"]))
    total_points -= goals_conceded_points
    
    yellow_card_points = calculate_yellow_card_points(int(row["yellow_cards"]))
    total_points -= yellow_card_points
    
    red_card_points = calculate_red_card_points(int(row["red_cards"]))
    total_points -= red_card_points
    
    own_goal_points = calculate_own_goal_points(int(row["own_goals"]))
    total_points -= own_goal_points
    
    return total_points
```

```python
def calculate_played_points(mins):
    if mins >= 60:
        return 2
    elif mins == 0:
        return 0
    else:
        return 1

def calculate_goals_points(goals,pos):
    if pos == "F":
        return 4*goals
    elif pos == "M":
        return 5*goals
    else:
        return 6*goals
    
def calculate_assist_points(assists):
    return assists*3
    
def calculate_clean_sheet_points(clean_sheet,pos):
    if pos == "G" or pos == "D":
        if clean_sheet == "True":
            return 4
        else:
            return 0
    elif pos == "M":
        if clean_sheet == "True":
            return 1
        else:
            return 0
    else:
        return 0
    
def calculate_pen_save_points(saves,pos):
    if pos == "G":
        return saves*5
    else:
        return 0
    
def calculate_pen_miss_points(misses):
    return misses*2
    
def calculate_goals_conceded_points(goals_conceded,pos):
    if pos == "G" or pos == "D":
        points = int(goals_conceded/2)
        return points
    else:
        return 0
    
def calculate_yellow_card_points(cards):
    return cards
    
def calculate_red_card_points(cards):
    return cards*3
    
def calculate_own_goal_points(own_goals):
    return own_goals*2
```

```python
df["points"] = df.apply(calculate_fpl_points,axis=1)
```

```python
df.to_csv("data/all_fpl_points.csv",index=False)
```
